```python
##### Load best performing motion models #####
symmetric_model = models.load_model('../saved models/symmetric/red10-64-b1/model.h5', custom_objects={'PSNR': PSNR})
random_model =  models.load_model('../saved models/random/red10-64-b1/model.h5', custom_objects={'PSNR': PSNR})
mixed_model = models.load_model('../saved models/mixed/red10-32-b1/model.h5', custom_objects={'PSNR': PSNR})
```

```python
#### Load Motion val Data###
f = h5py.File("../data/motion_data.hdf5", "r") 
motion_val_data = f.get("val_data").value
f.close()
```

```python
### Prepare Data and predict ####
images_clean = motion_val_data[:25]
images_noised_symmetric = add_motion_noise_to_images_and_add_dim(images_clean, motion_type = 'symmetric')
images_noised_random =add_motion_noise_to_images_and_add_dim(images_clean, motion_type = 'random')

output_mixedmodel_randnoise = mixed_model.predict(images_noised_random)
output_mixedmodel_symnoise = mixed_model.predict(images_noised_symmetric)

```

```python
#### calculate PSNR between Groundtruth and Noised####
psnr_randnoise_mean= calculate_psnr_mean(images_clean, np.squeeze(images_noised_random))
psnr_symnoise_mean = calculate_psnr_mean(images_clean, np.squeeze(images_noised_symmetric))
print('psnr between ground truth and random noised: '+ str(psnr_randnoise_mean))
print('psnr between ground truth and symmetric noised: ' + str(psnr_symnoise_mean))
```

```python
#### calculate PSNR between Groundtruth and Denoised from mixed trained model####
psnr_mixedmodel_random = calculate_psnr_mean(images_clean, np.squeeze(output_mixedmodel_randnoise))
psnr_mixedmodel_symmetric = calculate_psnr_mean(images_clean, np.squeeze(output_mixedmodel_symnoise))
print('psnr between ground truth and random denoised: '+ str(psnr_mixedmodel_random))
print('psnr between ground truth and symmetric denoised: '+ str(psnr_mixedmodel_symmetric))
```

```python
## get low motion examples ##
path_low = '../data/motion-images/low-motion/'
low_motion_imgs = []
for i in range(5):
    img  = str(i + 1) + '.jpg'
    img = np.asarray(skimage.io.imread(path_low + img)[:,:,0])
    img =img.astype(float) /255.
    low_motion_imgs.append(img)
low_motion_imgs = np.array(low_motion_imgs)
```

```python
## get medium motion examples ##
path_medium = '../data/motion-images/medium-motion/'
medium_motion_imgs = []
for i in range(6):
    
    img = str(i + 1) + '.jpg'
    img = skimage.io.imread(path_medium + img)[:,:,0]
    img = img.astype(float) /255
    medium_motion_imgs.append(img)
medium_motion_imgs = np.array(medium_motion_imgs)
```

```python
## get high motion examples ##
path_high =  '../data/motion-images/medium-motion/'
high_motion_imgs = []
for i in range(5):
    
    img  = str(i + 1) + '.jpg'
    img = skimage.io.imread(path_high + img)[:,:,0]
    img = img.astype(float) /255.
    high_motion_imgs.append(img)
high_motion_imgs = np.array(high_motion_imgs)
```

```python
denoised_low_by_symmetric = symmetric_model.predict(add_dim(low_motion_imgs))
denoised_low_by_random = random_model.predict(add_dim(low_motion_imgs))
denoised_low_by_mixed = mixed_model.predict(add_dim(low_motion_imgs))
```

```python
denoised_medium_by_symmetric = symmetric_model.predict(add_dim(medium_motion_imgs))
denoised_medium_by_random = random_model.predict(add_dim(medium_motion_imgs))
denoised_medium_by_mixed = mixed_model.predict(add_dim(medium_motion_imgs))

   
```

```python
denoised_high_by_symmetric = symmetric_model.predict(add_dim(high_motion_imgs))
denoised_high_by_random = random_model.predict(add_dim(high_motion_imgs))
denoised_high_by_mixed = mixed_model.predict(add_dim(high_motion_imgs))
```

```python
#Calculate total PSNR of Models output with ground truth images
PSNR_sym = np.mean((calculate_psnr_mean(low_motion_imgs, np.squeeze(denoised_low_by_symmetric)),calculate_psnr_mean(medium_motion_imgs, np.squeeze(denoised_medium_by_symmetric)),calculate_psnr_mean(high_motion_imgs, np.squeeze(denoised_high_by_symmetric))))
PSNR_rand = np.mean((calculate_psnr_mean(low_motion_imgs, np.squeeze(denoised_low_by_random)),calculate_psnr_mean(medium_motion_imgs, np.squeeze(denoised_medium_by_random)),calculate_psnr_mean(high_motion_imgs, np.squeeze(denoised_high_by_random))))
PSNR_mixed = np.mean((calculate_psnr_mean(low_motion_imgs, np.squeeze(denoised_low_by_mixed)),calculate_psnr_mean(medium_motion_imgs, np.squeeze(denoised_medium_by_mixed)),calculate_psnr_mean(high_motion_imgs, np.squeeze(denoised_high_by_mixed))))

print('Symmetric PSNR mean: ' + str(PSNR_sym))
print('Random PSNR mean: ' + str(PSNR_rand))
print('Mixed PSNR mean: ' + str(PSNR_mixed))
```
